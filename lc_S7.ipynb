```python
import math
import re
total = 1
firm_score = {}
nums = []
wanted_firm = raw_input('Enter firm for analysis:')


while total < 55:
   file_string1 = str(total)
   file_string2 = ".txt"
   file = file_string1 + file_string2
   fhand = open(file)
   text = fhand.read()

   firmy = re.findall('(.+):\sThe Legal Cheek trainee', text)
   firm = firmy[0]

   question = "Q. How good are your law firm's perks?"
   listy = re.findall('perks.\s\S+\s\S+\s([0-9]+)', text)

   ns = []
   for item in listy:
       try:
          inty = float(item)
          #print item
          ns.append(inty)
          nums.append(inty)
       except:
           #print item
           continue

   t = len(ns)
   average = sum(ns)/t
   

   firm_score[firm] = average
   total = total + 1

n = len(nums)
global_average = sum(nums)/n

list2 = []


for item in nums:
    mnums = (item - global_average)**2
    list2.append(mnums)

snums = sum(list2)/(n-1)
global_stdev = math.sqrt(snums)
global_st_err = global_stdev/math.sqrt(n)


firm_info = list()
for k, v in firm_score.items():
    firm_info.append( (v, k) )


firm_info.sort(reverse = True)

firm_name = []
firm_score = []

for v, k in firm_info:
   
   firm_name.append(k)
   firm_score.append(v)



total = 0

graph_y = []
y_firm = []
x_firm = []



tt = 1

for item in firm_name:
    #print item
    if item != wanted_firm:
        #firm_name.remove(item)
        graph_y.append(tt)
        tt = tt + 1
    if item == wanted_firm:
        pos = firm_name.index(item)
        y_firm.append(tt)
        tt = tt + 1

x_firm.append(firm_score[pos])
firm_score.remove(firm_score[pos])


graph_x = firm_score

print graph_y
print graph_x

print y_firm
print x_firm

print "Fig. 7.2: The average score out of 10 for " + wanted_firm + " and relative ranking compared to other firms\
 in the survey (" + question + ")."
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Bar(
    y=graph_y,
    x=graph_x,
    orientation = 'h'
)
trace2 = go.Bar(
    y=y_firm,
    x=x_firm,
    orientation = 'h'
)
data = [trace1, trace2]
layout = go.Layout(
    showlegend=False,
    xaxis=dict(
        range=[0, 10],
        title='Average score out of 10'
    ),
    yaxis=dict(
        range=[55, 0],
        title='Ranking'
    )
)


fig = go.Figure(data=data, layout=layout)
py.plot(fig)
py.image.save_as(fig, filename= wanted_firm + '_S7_Fig2.png')

#fig = go.Figure(data=data, layout=layout)
#py.plot(fig)
```

```python
import math
import re
total = 1
firm_score = {}
firm_err = {}
comps = []
comparables = []
nums = []
#Firm to analyse
question = "Q. How good are your law firm's perks?"
wanted_firm = raw_input('Enter firm for analysis:')

#set-up firms in peer group
if wanted_firm == 'Freshfields':
    comp = ['Allen & Overy', 'Slaughter and May', 'Linklaters', 'Clifford Chance']

   
print 'Peer Firms:'

#loop through all files, open and read
while total < 55:
   file_string1 = str(total)
   file_string2 = ".txt"
   file = file_string1 + file_string2
   fhand = open(file)
   text = fhand.read()

   #locate the firm name
   firmy = re.findall('(.+):\sThe Legal Cheek trainee', text)
   firm = firmy[0]


   #locate the individual responses for the question and make a list
   listy = re.findall('perks.\s\S+\s\S+\s([0-9]+)', text)


   #make a list of individual responses for the 'peer firms', to later calculate mean, st.err. etc.
   if firm in comp:
       print firm
       for item in listy:
           comps.append(item)

   ns = []
   #print firm
   #Loop through all responses, convert to a float and append to 2 lists (ns and nums)
   for item in listy:
       #print firm
       #print item
       try:
          inty = float(item)
          #print item
          ns.append(inty)
          nums.append(inty)
       except:
           #print item
           continue


   #print total
   #print firm
   #for item in listy:
       #print item

   # calculate the sample number for all numbers
   t = len(ns)

   #calculate the average for all numbers
   average = sum(ns)/t


   #calculate the standard deviation and standard error for all firms
   list_1 = []

   for item in ns:
       mnums1 = (item - average)**2
       list_1.append(mnums1)

   snums1 = sum(list_1)/(t-1)
   firm_stdev = math.sqrt(snums1)
   firm_st_err = firm_stdev/math.sqrt(t)

   #creates a dictionary containing the averages for all firms
   firm_score[firm] = average

   #creates a dictionary containing the standard errors for all firms
   firm_err[firm] = firm_st_err

   total = total + 1



n = len(nums)
global_average = sum(nums)/n

list2 = []


#calculating standard dev and err.

for item in nums:
    #print item
    mnums = (item - global_average)**2
    list2.append(mnums)

snums = sum(list2)/(n-1)
global_stdev = math.sqrt(snums)
global_st_err = global_stdev/math.sqrt(n)

#print nums
#print n

#For list of peer raw values, convert to float and append to new list: 'comparables'
for item in comps:
    int_item = float(item)
    comparables.append(int_item)

#work out length(n) and mean of comparables list
n_comp = len(comparables)
average_comp = sum(comparables)/n_comp


#calculating standard dev and err. of comparables list
list_comp = []

for item in comparables:
    #print item
    nums_comp = (item - average_comp)**2
    list_comp.append(nums_comp)

snums_comp = sum(list_comp)/(n_comp - 1)
stdev_comp = math.sqrt(snums_comp)
st_err_comp = stdev_comp/math.sqrt(n_comp)

#Returns the mean score for the 'chosen firm'
chosen_firm = firm_score[wanted_firm]

#Returns the mean error for the 'chosen firm'
chosen_err = firm_err[wanted_firm]

#VALUES TO USE FOR BAR GRAPH CONSTRUCTION

x_graph = [wanted_firm, "Peers", "All firms"]
print x_graph

y_graph = [chosen_firm, average_comp, global_average]
print y_graph

err_graph = [chosen_err, st_err_comp, global_st_err]
print err_graph

print "Fig. 7.3: The average score out of 10 for " + wanted_firm + " (" + str(chosen_firm)[:4] + ") compared with average scores out of 10 for\
 'peer firms'" + " (" + str(average_comp)[:4] + ") and 'all firms'" + " (" + str(global_average)[:4] + ") (" + question + "). Error bars represent\
 the standard error of all responses."

lower_range_y = 4
upper_range_y = 9
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

trace1 = go.Bar(
    x=x_graph,
    y=y_graph,
    error_y=dict(
               type='data',
               array=err_graph,
               visible=True
               )
)

data = [trace1]

layout = go.Layout(
    yaxis=dict(
        range=[4, 9],
        title ='Average score out of 10'
    
    )
)

fig = go.Figure(data=data, layout=layout)
py.plot(fig)
py.image.save_as(fig, filename= wanted_firm + '_S7_Fig3.png')
```
